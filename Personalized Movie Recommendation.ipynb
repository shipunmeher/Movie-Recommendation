# Converting Text to Feature Vectors

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
vect_features = vectorizer.fit_transform(df["genres"])
print(vect_features)
```

# Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vect_features)
print(similarity)
```

```python
similarity.shape
```

```python
title_list = df["title"].tolist()
title_list
```

# Getting MOVIE name from User

```python
movie_name = input("Enter your Movie name You Want :")
```

Finding Close Match Movie for User

```python
import difflib
movie_recommend = difflib.get_close_matches(movie_name,title_list)
movie_recommend
```

```python
close_movie_recommend = movie_recommend[0]
print(close_movie_recommend)
```

# Finding Index No. Of Movie

```python
movie_index = df[df.title==close_movie_recommend]['movieId'].values[0]
movie_index
```

```python
similarity_score = list(enumerate(similarity[movie_index]))
len(similarity_score)
```

Sorting Movies Based On Similarity Score

```python
similar_movies_sorted = sorted(similarity_score,key=lambda x:x[1],reverse=True)
```

# Printing Similary Movie

```python
print("Movies Recommend For You")
i = 1
for movie in similar_movies_sorted:
    index = movie[0]
    title_row = df[df.movieId == index]["title"]
    if not title_row.empty:
        title_from_index = title_row.values[0]
        if i <= 10:
            print(i, ":-", title_from_index)
        i += 1
```
